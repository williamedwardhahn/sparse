```python
def plot(x):
    fig, ax = plt.subplots()
    im = ax.imshow(x)
    ax.axis('off')
    fig.set_size_inches(18, 10)
    plt.show()
```

```python
def montage(X):    
    m, n, count = np.shape(X)    
    mm = int(np.ceil(np.sqrt(count)))
    nn = mm
    M = np.zeros((mm * m, nn * n))

    image_id = 0
    for j in range(mm):
        for k in range(nn):
            if image_id >= count: 
                break
            sliceM, sliceN = j * m, k * n
            M[sliceN:sliceN + n, sliceM:sliceM + m] = X[:, :, image_id]
            image_id += 1
    return M
```

```python
def max_pool(x):
    block_shape = 2,2,3
    blocks = view_as_windows(x, block_shape)
    blocks_vec = blocks.reshape(blocks.shape[0], blocks.shape[1], -1)
    max_pool = np.max(blocks_vec, axis=2)
    return max_pool
```

```python
s = [72,72]
p = [9,9]    #patch size
n = 9          #number of receptive fields
c = 1          #color channels
b = 256         #batch size
```

```python
def convolve(img):
    with tf.Graph().as_default():
        

        x = tf.placeholder('float32', [b, s[0], s[1], c])
        
        
        w1 = tf.Variable(tf.random_normal([p[0], p[1], n, c]))
        w2 = tf.Variable(tf.random_normal([2*p[0], 2*p[1], n, c]))
        
        
        a1 = tf.Variable(tf.random_normal([b, s[0], s[1], n]))
        a2 = tf.Variable(tf.random_normal([b, s[0], s[1], n]))
        

        
        
        y1 = tf.nn.conv2d(a1, w1, strides=[1, 1, 1, 1], padding='SAME')
        y2 = tf.nn.conv2d(a2, w2, strides=[1, 1, 1, 1], padding='SAME')
        
        
        
        
        y1 = tf.reduce_sum(y1, 3, keep_dims=True)
        y2 = tf.reduce_sum(y2, 3, keep_dims=True)  
        
        y = y1 + y2
        print(y1.shape)
        print(y2.shape)
        print(y.shape)
        
        z = tf.subtract(x,y)

        
        loss = tf.norm(z, 2) + 0.01*tf.norm(w1, 2) + 0.01*tf.norm(w2, 2) + 0.1*tf.norm(a1, 1) + 0.1*tf.norm(a2, 1)
        
        
        
        

        optimizer = tf.train.AdamOptimizer(learning_rate)
        
        
        train_w = optimizer.minimize(loss, var_list=[w1,w2])
        
        train_a = optimizer.minimize(loss, var_list=[a1,a2])
        
        
        init = tf.global_variables_initializer()
        
        with tf.Session() as session:
        
            session.run(init)
            
            batch = np.random.randint(0,img.shape[0],b)  
            # Training cycle
            for epoch in range(training_epochs):
                
#                 for i in range(10):
                
                

                session.run([train_a],feed_dict={x: img[batch,:,:,:]})

#                 for i in range(100):

#                     batch = np.random.randint(0,img.shape[0],b) 

                session.run([train_w],feed_dict={x: img[batch,:,:,:]})
        

                print(epoch)
            
            
            L = loss.eval(feed_dict={x: img[batch,:,:,:]})
          
            y = y.eval()
            
            w1 = w1.eval()
            a1 = a1.eval()
            w2 = w2.eval()
            a2 = a2.eval()
            
        return L,y,w1,w2,a1,a2
```

```python
# tf Graph input
y = tf.placeholder("float", [M,1])
D = tf.placeholder("float", [M,N])
x = tf.Variable(tf.random_normal([N,1]))
```

```python
# Create model
def CS(D,x):
    
    f = tf.subtract(y,tf.matmul(D,x))
    f = tf.norm(f, ord=2)
    f = tf.add(f, tf.norm(x, ord=1))
    
    return f
```

```python
# Construct model
f = CS(D,x)
```

```python
learning_rate = 0.01
training_epochs = 1000
```

```python
# Define loss and optimizer
cost = f
optimizer = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost)
```

```python
# Initializing the variables
init = tf.global_variables_initializer()
```

```python
# Launch the graph
with tf.Session() as sess:
    
    sess.run(init)

    for epoch in range(training_epochs):
      
        sess.run([optimizer, cost], feed_dict={y: y1, D: D1})
    
    x3=x.eval()
```
